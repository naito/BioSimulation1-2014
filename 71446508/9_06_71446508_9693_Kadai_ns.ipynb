###cdc2関連

```python
R4['k4'] = 180
R6['k6'] = 1
plt.figure()
plt.title( 'cdc2 dynamics' )
for FullPN, d in Data_dict.items():
    if FullPN in [ CP, pM, M ]:
        plt.plot( d[ :, 0 ], d[ :, 1 ], label = FullPN.split(':')[ 2 ] )

plt.legend( loc = 'upper right' )
```

pMは徐々に蓄積していきおよそ0.3に達したところで大部分がMに変化するために急減する。CPはpMに変化していくためpMの増加とともに減少する。
R4の値を減らすとMへの変化速度に対してC2への変化速度の方が大きくなっていくため、pMがほぼ0となる前にR6~R8~R3とサイクルがまわってまta
pMが蓄積し始める。（下図

```python
cdc2_k4_and_k6(94,1)
```

R4の値をさらに減らすとMへの変化速度に対してC2への変化速度がさらに大きくなっていくため、Mは生成されてもすぐに分解されてしまうため、ほとんどたまらなくなっていく。

```python
cdc2_k4_and_k6(90,1)
```

同様にしてR4をさらに減らすと、上の図からも推測できるように、グラフが直線に近づいていき、やがて定常状態になる。

```python
cdc2_k4_and_k6(80,1)
```

逆にR6の値を大きくしていっても、Mの分解速度が大きくなってMがたまりにくくなり、グラフは直線へ近づいていく。

```python
cdc2_k4_and_k6(180,1.45)
```
